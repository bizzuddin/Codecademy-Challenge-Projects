# U.S. Medical Insurance Costs

```python
import csv
from decimal import Decimal
```

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
```

```python
with open('insurance.csv') as insurance_data:
    ins_data_dict = csv.DictReader(insurance_data)
    ins_data_list = []    
    for data in ins_data_dict:
        ins_data_list.append(data)
    for item in ins_data_list:
        age.append(item['age'])
        sex.append(item['sex'])
        bmi.append(item['bmi'])
        children.append(item['children'])
        smoker.append(item['smoker'])
        region.append(item['region'])
        charges.append(item['charges'])
```

```python
class InsuranceAnalysis:
    def __init__(self,age,sex,bmi,children,smoker,region,charges):
        self.pat_age = age
        self.pat_sex = sex
        self.pat_bmi = bmi
        self.pat_children = children
        self.pat_smoker = smoker
        self.pat_region = region
        self.pat_charges = charges
        self.pat_total = len(age)
    
    def overall_charges_data(self):
        pass
        
    def age_charge_analysis(self):
        total_age = 0
        i = 0
        under_twenty = 0
        twenty_count = 0
        twenty_thirty = 0
        thirty_count = 0
        thirty_forty = 0
        forty_count = 0
        forty_fifty = 0
        fifty_count = 0
        fifty_sixty = 0
        sixty_count = 0
        above_sixty = 0
        abv_sixty_count = 0
        total_cunt = self.pat_total
        for age in self.pat_age:
            int_age = int(age)
            total_age += int_age
            if int_age <= 20:
                under_twenty += float(self.pat_charges[i])
                twenty_count += 1
            elif int_age > 20 and int_age <= 30:
                twenty_thirty += float(self.pat_charges[i])
                thirty_count += 1
            elif int_age > 30 and int_age <= 40:
                thirty_forty += float(self.pat_charges[i])
                forty_count += 1
            elif int_age > 20 and int_age <= 50:
                forty_fifty += float(self.pat_charges[i])
                fifty_count += 1
            elif int_age > 20 and int_age <= 60:
                fifty_sixty += float(self.pat_charges[i])
                sixty_count += 1
            elif int_age > 60:
                above_sixty += float(self.pat_charges[i])
                abv_sixty_count += 1
            i+=1            
        avg_age = round(total_age/total_cunt,2)
        avg_twenty = round(under_twenty/twenty_count,2)
        avg_thirty = round(twenty_thirty/thirty_count,2)
        avg_forty = round(thirty_forty/forty_count,2)
        avg_fifty = round(forty_fifty/fifty_count,2)
        avg_sixty = round(fifty_sixty/sixty_count,2)
        avg_abv_sixty = round(above_sixty/abv_sixty_count,2)
        age_sen = """ 
Average age: {} years old
Patients by age (average insurance cost)
<20: {} (${})
21-30: {} (${})
31-40: {} (${})
41-50: {} (${})
51-60: {} (${})
>60: {} (${})
""".format(avg_age,twenty_count,avg_twenty,thirty_count,avg_thirty,forty_count,avg_forty,fifty_count,avg_fifty,sixty_count,avg_sixty,abv_sixty_count,avg_abv_sixty)
        print(age_sen)
        
    def sex_charges_analysis(self):
        sex_list = self.pat_sex
        fem = 0
        fem_charge = 0
        mal = 0
        mal_charge = 0
        i=0
        for sex in sex_list:
            if sex == 'female':
                fem+=1
                fem_charge += float(self.pat_charges[i])
            elif sex == 'male':
                mal+=1
                mal_charge += float(self.pat_charges[i])
            i+=1
        fem_per = round(100*fem/(fem+mal),2)
        mal_per = round(100*mal/(fem+mal),2)
        avg_fem = round(fem_charge/fem,2)
        avg_mal = round(mal_charge/mal,2)
        
        sex_sen = """
Patients by gender
Female: {} ({}%)
Male: {} ({}%)
Average insurance cost
Female: ${}
Male: ${}
        """.format(fem,fem_per,mal,mal_per,avg_fem,avg_mal)
        print(sex_sen)
    
    def bmi_charges_analysis(self):
        total_bmi = 0.0
        i = 0
        #underweight < 18.5
        under_cnt = 0
        under_crg = 0
        #normal = 18.5-24.9
        normal_cnt = 0
        normal_crg = 0
        #overweight = 25-29.9
        over_cnt = 0 
        over_crg = 0
        #obese > 30
        obs_cnt = 0
        obs_crg = 0
        for bmi in self.pat_bmi:
            flt_bmi = float(bmi)
            if flt_bmi < 18.5:
                under_cnt += 1
                under_crg += float(self.pat_charges[i])
            elif flt_bmi >= 18.5 and flt_bmi < 25.0:
                normal_cnt += 1 
                normal_crg += float(self.pat_charges[i])
            elif flt_bmi >= 25.0 and flt_bmi < 30.0:
                over_cnt += 1
                over_crg += float(self.pat_charges[i])
            elif flt_bmi >= 30.0:
                obs_cnt += 1
                obs_crg += float(self.pat_charges[i])
            i+=1
        avg_bmi = round(total_bmi/self.pat_total,2)
        avg_under = round(under_crg/under_cnt,2)
        avg_normal = round(normal_crg/normal_cnt,2)
        avg_over = round(over_crg/over_cnt,2)
        avg_obs = round(obs_crg/obs_cnt,2)
        bmi_sen = """
Average BMI: {}
Patients by BMI (Average insurance cost)
Underweight (<18.5): {} (${})
Normal (18.5-24.9): {} (${})
Overweight (25-29.9): {} (${})
Obese (>30): {} (${})
        """.format(avg_bmi,under_cnt,avg_under,normal_cnt,avg_normal,over_cnt,avg_over,obs_cnt,avg_obs)
        print(bmi_sen)
        
    def children_charge_analysis(self):
        child_list = self.pat_children
        i=0
        zero_child = 0
        one_child = 0
        two_child = 0
        three_child = 0
        four_child = 0
        fivenmore = 0
        zero_charge = 0
        one_charge = 0
        two_charge = 0
        three_charge = 0
        four_charge = 0
        fivenmore_charge = 0
        for child in child_list:
            child_num = int(child)
            if child_num == 0:
                zero_child += 1
                zero_charge+= float(self.pat_charges[i])
            elif child_num == 1:
                one_child += 1
                one_charge+= float(self.pat_charges[i])
            elif child_num == 2:
                two_child += 1
                two_charge+= float(self.pat_charges[i])
            elif child_num == 3:
                three_child += 1
                three_charge+= float(self.pat_charges[i])
            elif child_num == 4:
                four_child += 1
                four_charge+= float(self.pat_charges[i])
            elif child_num >= 5:
                fivenmore += 1
                fivenmore_charge+= float(self.pat_charges[i])
            i+=1
        #print(total_children,zero_child,one_child,two_child,three_child,four_child,five_child,sixnmore)
        zero_avg = round(zero_charge/zero_child,2)
        one_avg = round(one_charge/one_child,2)
        two_avg = round(two_charge/two_child,2)
        three_avg = round(three_charge/three_child,2)
        four_avg = round(four_charge/four_child,2)
        fivenmore_avg = round(fivenmore_charge/fivenmore,2)
        child_info = """
Patients by their number of children (Average insurance cost)
0: {} (${})
1: {} (${})
2: {} (${})
3: {} (${})
4: {} (${})
5: {} (${})
        """.format(zero_child,zero_avg,one_child,one_avg,two_child,two_avg,three_child,three_avg,four_child,four_avg,fivenmore,fivenmore_avg)
        print(child_info)
        
    def smoker_charge_analysis(self):
        smoker_cnt = 0
        smoker_crg = 0
        non_cnt = 0
        non_crg = 0
        i = 0
        for status in self.pat_smoker:
            if status == 'yes':
                smoker_cnt +=1
                smoker_crg += float(self.pat_charges[i])
            elif status == 'no':
                non_cnt +=1
                non_crg += float(self.pat_charges[i])
            i+=1
        smoker_avg = round(smoker_crg/smoker_cnt,2)
        non_svg = round(non_crg/non_cnt,2)
        smoker_sen ="""
Patients by smoker status (Average insurance cost)
Smoker: {} (${})
Non-smoker: {} (${})
""".format(smoker_cnt,smoker_avg,non_cnt,non_svg)
        print(smoker_sen)
    def region_analysis(self):
        regions = []
        for region in self.pat_region:
            if region not in regions:
                regions.append(region)            
        tot_region = len(regions)
        region_sen = "There are {} regions that divide patients residence. Those regions are:".format(tot_region)
        print(region_sen,regions)
    def dict_maker(self):
        patients_dict = {}
        patients_dict['Age'] = self.pat_age
        patients_dict['Sex'] = self.pat_sex
        patients_dict['BMI'] = self.pat_bmi
        patients_dict['Children'] = self.pat_children
        patients_dict['Smoker'] = self.pat_smoker
        patients_dict['Region'] = self.pat_region
        patients_dict['Charges'] = self.pat_charges
        return patients_dict
        
    
```
